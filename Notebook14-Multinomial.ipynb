# Handout 14: Multinomial Distributions

## Exer 14.1

```python
P = BoxModel(['R', 'I', 'D', 'O'], probs = [0.28, 0.43, 0.27, 0.02], size=20)
P.sim(5)
```

```python
X_R = RV(P, count_eq('R'))
X_I = RV(P, count_eq('I'))
X_D = RV(P, count_eq('D'))
X_O = RV(P, count_eq('O'))
X = (X_R & X_I & X_D & X_O)
```

```python
outcome = P.draw()
outcome
```

```python
X_R(outcome), X_I(outcome), X_D(outcome), X_O(outcome), X(outcome)
```

```python
x = X.sim(100000)
x
```

## a)

```python
def event(x):
    if (x[0] == 6) & (x[1] == 9) & (x[2] == 5) & (x[3] == 0):
        return True
    
x.count(event) / 100000
```

```python
( (X_R == 6) & (X_I == 9) & (X_D == 5) & (X_O == 0) ).sim(100000).tabulate()
```

## c)

```python
x_R = X_R.sim(10000)
x_R.plot()
Binomial(n = 20, p = 0.28).plot()
x_R.mean(), x_R.var(), x_R.count_eq(6) / 10000
```

## d)

```python
x_RD = (X_R + X_D).sim(10000)
x_RD.plot()
Binomial(n = 20, p = 0.28 + 0.27).plot()
x_RD.mean(), x_RD.var(), x_RD.count_eq(11) / 10000
```

## e)

```python
x_R_given_xDeq5 = (X_R | (X_D == 5) ).sim(1000)
x_R_given_xDeq5.plot()
Binomial(n = 20 - 5, p = 0.28 / (1 - 0.27)).plot()
x_R_given_xDeq5.mean(), x_R_given_xDeq5.var(), x_R_given_xDeq5.count_eq(6) / 1000
```
