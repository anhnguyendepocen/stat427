## Exer 7.1

### a) Most powerful level $\alpha_0=0.05$ test for $n=10000$

```python
alpha0 = 0.05

n = 10000

theta0 = 0.5

c = 1 + Binomial(n, theta0).quantile(1 - alpha0)

size = 1 - Binomial(n, theta0).cdf(c - 1)

Binomial(n, theta0).plot()

plt.xlim(n * theta0 - 3 * sqrt(n * theta0 * (1 - theta0)),
         n * theta0 + 3 * sqrt(n * theta0 * (1 - theta0)))
plt.legend([r'Null distribution of $X$']);
plt.axvline(x = c, color = 'orange');
plt.axvspan(c, n, alpha = 0.5, color = 'gray');
plt.xlabel(r"$x$", fontsize=18)
plt.ylabel(r"$p(x)$", fontsize=18);
plt.title(r"Reject $H_0:\theta =%s$" % theta0 +
          r" if $X\geq %s$" % c +
          r", for $X\sim$ Binomial$(%s, \theta)$" % n +
          "\n Size = %s" % round(size, 4));
```

## Exer 7.4

### p-value for $\bar{x} = 97.5$

```python
x = 97.5

n = 4

sigma = 1

theta0 = 98.6

c = abs(x - theta0)

c1 = theta0 - c
c2 = theta0 + c

size = 2 * Normal(theta0, sigma / sqrt(n)).cdf(c1)

Normal(theta0, sigma / sqrt(n)).plot()

plt.xlim(theta0 - 3 * sigma / sqrt(n), theta0 + 3 * sigma / sqrt(n))
plt.legend([r'Null distribution of $\bar{X}$']);
plt.axvline(x = c1, color = 'orange');
plt.axvline(x = c2, color = 'orange');
plt.axvspan(90, c1, alpha = 0.5, color = 'gray');
plt.axvspan(c2, 110, alpha = 0.5, color = 'gray');
plt.xlabel(r"$\bar{x}$", fontsize=18)
plt.ylabel(r"$p(\bar{x})$", fontsize=18);
plt.title(r"Reject $H_0:\mu =%s$" % theta0 +
          r" if $\bar{X}\leq %s$" % round(c1, 1) +
          r" or $\bar{X}\geq %s$" % round(c2, 1) +
          r", for $\bar{X}\sim N(\mu, %s$" % sigma +
          r" $/ \sqrt{%s}$)" % n + 
          "\n p-value = %s" % round(size, 4));
```

### d) Null distribution of p-value

```python
mu0 = 98.6

n = 4

sigma = 1

P0 = Normal(mu0, sigma) ** n

def pvalue_x(x):
    z = abs(x.mean() - mu0) / (sigma / sqrt(n))
    return 2 * Normal(0, 1).cdf(-z)

Pvalue_H0 = RV(P0, pvalue_x)

pvalues_H0 = Pvalue_H0.sim(10000)

pvalues_H0.plot()

Uniform(0, 1).plot()
```

### Alternative distribution of p-value

```python
mu0 = 98.6

mu1 = 97.5

n = 4

sigma = 1

P1 = Normal(mu1, sigma) ** n # simulate under H1

def pvalue_x(x):
    z = abs(x.mean() - mu0) / (sigma / sqrt(n)) # standardize wrt H0
    return 2 * Normal(0, 1).cdf(-z)

Pvalue_H1 = RV(P1, pvalue_x)

pvalues_H1 = Pvalue_H1.sim(10000)

pvalues_H1.plot()
```
