# Handout 11: Power of $t$ tests and effect size

## Exer 11.1

```python
n = 9
mu0 = 0
mu = 1
sigma = 1

P = Normal(mu, sigma) ** n
Xbar = RV(P, mean)
S2 = RV(P, var) * (n / (n - 1))
S = sqrt(S2)

T = (Xbar - mu0) / (S / sqrt(n))

t = T.sim(10000)
t.plot()
```

## Noncentral t sim

```python
Z, V = RV(Normal(0, 1) * ChiSquare(n - 1))

psi = sqrt(n) * (mu - mu0) / sigma

U = (Z + psi) / sqrt(V / (n - 1))

u = U.sim(10000)

u.plot()
```

## Distribution of $T$ statistic

```python
t.plot()

u.plot('density')
```

## Null versus alternative distribution of $T$ statistic

```python
StudentT(n - 1).plot()
u.plot('density')

plt.xlim(-5, 9)
plt.legend([r'Null distribution of $T$',
            r'Alternative distribution of $T$ if $\mu=%s$' % mu]);
plt.axvline(x = -2, color = 'orange');
plt.axvline(x = 2, color = 'orange');
plt.axvspan(-30, -2, alpha = 0.5, color = 'gray');
plt.axvspan(2, 30, alpha = 0.5, color = 'gray');
```
