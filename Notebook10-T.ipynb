# Handout 10: T tests

```python
n = 4

StudentT(n - 1).plot()

alpha0 = 0.05

t = StudentT(n - 1).quantile(1 - alpha0 / 2)

plt.vlines(-abs(t), 0, 1, color = 'orange', alpha=.6, linewidth=4)
plt.axvspan(-20, -abs(t), alpha = 0.5, color = 'gray');
plt.vlines(abs(t), 0, 1, color = 'orange', alpha=.6, linewidth=4)
plt.axvspan(abs(t), 20, alpha = 0.5, color = 'gray');

t
```
