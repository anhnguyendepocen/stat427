# Handout 12: $F$ Distributions

```python
d1 = 10
d2 = 6

V1, V2 = RV(ChiSquare(d1) * ChiSquare(d2))

U = (V1 / d1) / (V2 / d2)

u = U.sim(10000)
u.plot()
```

```python
u.plot(bins = 1000)
plt.xlim(0, 10);
```

```python
u.plot(bins = 1000)
F(d1, d2).plot()
plt.xlim(0, 10);
```
